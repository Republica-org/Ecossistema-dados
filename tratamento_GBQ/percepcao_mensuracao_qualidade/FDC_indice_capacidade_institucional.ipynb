```python
client = bigquery.Client()
```

```python
dataset_ref = client.dataset('percepcao_mensuracao_qualidade_estado')
```

```python
df.info()
```

```python
schema = [bigquery.SchemaField('sigla_uf', 'STRING', description= 'Sigla da Unidade da Federação referente aquela observação'),
          bigquery.SchemaField('indice_capacidade_institucional', 'FLOAT', description= 'Índice de Capacidade Institucional (ICI), formado por dois elementos: Qualidade das Instituições e Qualidade da Gestão Pública'),
          bigquery.SchemaField('qualidade_institucional', 'FLOAT', description= 'Refere-se a Qualidade da Instituição, formada por quatro parâmetros: Qualidade do Judiciário, Segurança Pública, Controle da Corrupção, Qualidade Fiscal'),
          bigquery.SchemaField('qualidade_judiciario', 'FLOAT', description= 'Refere-se a Qualidade do Judiciário, subelemento de qualidade institucional'),
          bigquery.SchemaField('seguranca_publica', 'FLOAT', description= 'Refere-se a Segurança Pública,subelemento de qualidade institucional'),                    
          bigquery.SchemaField('qualidade_fiscal', 'FLOAT', description= 'Refere-se a Qualidade Fiscal, subelemento de qualidade institucional'),                    
          bigquery.SchemaField('controle_corrupcao', 'FLOAT', description= 'Refere-se ao Controle da Corrupção, subelemento de qualidade institucional'),                    
          bigquery.SchemaField('qualidade_gestao', 'FLOAT', description= 'Refere-se a Qualidade da Gestão Pública, formado por três parâmetros: Governo Aberto, Ambiente de Negócios e Diversidade'),
          bigquery.SchemaField('governo_aberto', 'FLOAT', description= 'Refere-se a Governo Aberto, subelemento de qualidade da gestão pública'),
          bigquery.SchemaField('ambiente_negocios', 'FLOAT', description= 'Refere-se a Ambiente de Negócios, subelemento de qualidade da gestão pública'),
          bigquery.SchemaField('diversidade_contratacao_publica', 'FLOAT', description= 'Refere-se a Diversidade na Contratação Pública, subelemento de qualidade da gestão pública')]
```

```python
table_ref = dataset_ref.table('FDC_capacidades_institucionais')
job_config = bigquery.LoadJobConfig(schema=schema)
job = client.load_table_from_dataframe(df, table_ref, job_config=job_config)
job.result() 
```
