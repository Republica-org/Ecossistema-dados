Subindo para o GBQ

```python
client = bigquery.Client()
dataset_ref = client.dataset('perfil_remuneracao')
```

```python
schema= [bigquery.SchemaField('ano','INTEGER',description='Ano de referência da observação'),
          bigquery.SchemaField('tipo_das','STRING',description='Nível de cargo de Direção e Assessoramento Superior (DAS)'),
          bigquery.SchemaField('taxa_crescimento','FLOAT', description='Taxa de crescimento da observação')]
```

```python
table_ref = dataset_ref.table('SIAPE_taxa_crescimento_das_nivel') 
job_config = bigquery.LoadJobConfig(schema=schema)
job = client.load_table_from_dataframe(df, table_ref, job_config=job_config)
job.result()
```
