```python
df.rename(columns={'perc_servidores_ano': 'ano', 'valor_percent':'per_valor'}, inplace=True)
df
```

```python
df['ministerios'] = df['ministerios'].str.title()
```

```python
def transformar(nome):
    nome = re.sub(r"\bMinisterio\b", "Ministério", nome)
    nome = re.sub(r"\bEducacao\b", "Educação", nome)
    nome = re.sub(r"\bSaude\b", "Saúde", nome)
    nome = re.sub(r"\bGestao\b", "Gestão", nome)
    nome = re.sub(r"\bServicos\b", "Serviços", nome)
    nome = re.sub(r"\bacoes\b", "ações", nome)
    nome = re.sub(r"\bJustica\b", "Justiça", nome)
    nome = re.sub(r"\bPecuaria\b", "Pecuária", nome)
    nome = re.sub(r"\bCiencia\b", "Ciência", nome)
    nome = re.sub(r"\bComercio\b", "Comércio", nome)
    nome = re.sub(r"\bIndustria\b", "Indústria", nome)
    nome = re.sub(r"\bAviacao\b", "Aviação", nome)
    nome = re.sub(r"\bRelacoes\b", "Relações", nome)
    nome = re.sub(r"\bInovacao\b", "Inovação", nome)
    nome = re.sub(r"\bComunicacoes\b", "Comunicações", nome)
    nome = re.sub(r"\bIntegracao\b", "Integração", nome)
    return nome
```

```python
df['ministerios']= df['ministerios'].apply(transformar)
df
```

```python
df['ministerios'].unique()
```

```python
df = df[['ano','ministerios', 'per_valor']]
```

Subindo para o GBQ

```python
client = bigquery.Client()
dataset_ref = client.dataset('perfil_remuneracao')
```

```python
schema=[bigquery.SchemaField('ano','INTEGER',description='Ano de referência da observação'),
 bigquery.SchemaField('ministerios','STRING',description='Ministérios observados'),
 bigquery.SchemaField('per_valor','FLOAT',description='Percentual referente ao numero de servidores')]
```

```python
table_ref = dataset_ref.table('SIAPE_civis_ativos_ministerio') # nome da tabela no padrão FONTE_algo_intuitivo_dado
job_config = bigquery.LoadJobConfig(schema=schema)
job = client.load_table_from_dataframe(df, table_ref, job_config=job_config)
job.result()  
```
