```python
import numpy as np
import pandas as pd
import numpy as np
import os
import time
from google.cloud import bigquery
import pandas_gbq
import glob
```

Pegando a tabela do BigQuery

```python
query1 = "select * from `repositoriodedadosgpsp.perfil_remuneracao.SIAPE_ativos_infraestrutura`"
```

```python
query2 = "select * from `repositoriodedadosgpsp.perfil_remuneracao.SIAPE_ativos_regulatoria`"
```

```python
query3 = "select * from `repositoriodedadosgpsp.perfil_remuneracao.SIAPE_ativos_social`"
```

```python
query4 = "select * from `repositoriodedadosgpsp.perfil_remuneracao.SIAPE_ativos_economia_rural`"
```

Transformando em DataFrame

```python
df1 = pandas_gbq.read_gbq(query1, project_id='repositoriodedadosgpsp')
```

```python
df1['area_ministerio']="Infraestrutura"
df1
```

```python
df2 = pandas_gbq.read_gbq(query2, project_id='repositoriodedadosgpsp')
```

```python
df2['area_ministerio'] = "Regulatória"
df2
```

```python
df3 = pandas_gbq.read_gbq(query3, project_id='repositoriodedadosgpsp')
```

```python
df3['area_ministerio'] = "Social"
df3
```

```python
df4 = pandas_gbq.read_gbq(query4, project_id='repositoriodedadosgpsp')
```

```python
df4['area_ministerio'] = "Economia Rural"
df4
```

```python
df = pd.concat([df1,df2,df3,df4])
df
```

```python
df = df[['ano','area_ministerio','ministerios', 'servidores_ativos_ministerios', 'per_servidores']]
df
```

```python
df['area_ministerio'].unique()
```

Subindo para o GBQ

```python
client = bigquery.Client()
dataset_ref = client.dataset('perfil_remuneracao')
```

```python
schema= [bigquery.SchemaField('ano','INTEGER',description='Ano de referência da observação'),
          bigquery.SchemaField('ministerios','STRING',description='Ministérios observados'),
          bigquery.SchemaField('servidores_ativos_ministerios', 'INTEGER',description='Número de servidores ativos por ministério'),
          bigquery.SchemaField('per_servidores','FLOAT', description='Percentual de servidores ativos por ministério '),
          bigquery.SchemaField('area_ministerio','STRING',description='Área em que os ministérios estão inclusos')]
```

```python
table_ref = dataset_ref.table('SIAPE_ativos_area_ministerios') 
job_config = bigquery.LoadJobConfig(schema=schema)
job = client.load_table_from_dataframe(df, table_ref, job_config=job_config)
job.result()
```
