```python
##importando pacotes
import pandas as pd
import numpy as np
import time
import os
import pandas_gbq
from google.cloud import bigquery
import glob
import openpyxl
import csv
import re
```

```python
df = pd.read_csv("indicador_PEP_5.csv",  decimal=',')
df
```

```python
df.info()
```

```python
# Reordenar as colunas conforme solicitado
df = df[['cargo_com_codigo', 'salario_inicial', 'salario_final', 'amplitude_salarial']]
```

```python
df
```

```python
print(df)
```

```python
schema=[
bigquery.SchemaField('cargo_com_codigo','STRING',description='Cargo e código de acordo com o site.'),
bigquery.SchemaField('salario_inicial','Float',description='Média de salario inicial do cargo'),
bigquery.SchemaField('salario_final','Float',description='Média de salario final do cargo'),
bigquery.SchemaField('amplitude_salarial','Float',description='Amplitude salarial do cargo')
]
```

```python
## Subindo para datalake
client = bigquery.Client(project='repositoriodedadosgpsp')
dataset_ref = client.dataset('estrutura_organizacional_carreiras')
```

```python
table_ref = dataset_ref.table('PEP_carreiras_amplitude_salarial_v2') # nome da tabela no padrão FONTE_algo_intuitivo_dado
job_config = bigquery.LoadJobConfig(schema=schema)
job = client.load_table_from_dataframe(df, table_ref, job_config=job_config)
job.result()
```
