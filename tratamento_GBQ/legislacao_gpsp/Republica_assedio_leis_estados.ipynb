```python
df = pd.read_csv("rju_legislacao_assedio.csv", sep=';')
```

```python
client = bigquery.Client()
dataset_ref = client.dataset('legislacoes')
```

```python
df=df[['UF', 'RJU', 'preve_assedio_proibicao_rju', 'lei_assedio_proibicao',
       'preve_assedio_proibicao_lei_esparsa']]
```

```python
df.columns
```

```python
schema=[bigquery.SchemaField('UF','STRING',description='Unidade da Federação.'),
bigquery.SchemaField('RJU','STRING',description='Qual lei institui o Regime Jurídico Único'),
bigquery.SchemaField('preve_assedio_proibicao_rju','STRING',description='Informa se existe a proibição de assédio no estatuto do servidor ou em lei esparsa'),
bigquery.SchemaField('lei_assedio_proibicao','STRING',description='Lei complementar que regula'),
bigquery.SchemaField('preve_assedio_proibicao_lei_esparsa','STRING',description='Qual lei esparsa proíbe.')
]
```

```python
table_ref = dataset_ref.table('Republica_assedio_leis_estados') # nome da tabela no padrão FONTE_algo_intuitivo_dado
job_config = bigquery.LoadJobConfig(schema=schema)
job = client.load_table_from_dataframe(df, table_ref, job_config=job_config)
job.result()
```
