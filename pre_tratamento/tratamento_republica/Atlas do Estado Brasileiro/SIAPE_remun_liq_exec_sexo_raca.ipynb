Separação da coluna "sexo_raça" em "Sexo" e "Raça"


```python
df['sexo_raça'].unique() #para mostrar os valores inclusos na coluna "sexo_raça"
```

```python
df['Sexo'] = df['sexo_raça'].apply(lambda x: 'Feminino' if x in ['Mulher Branca','Mulher Negra','Não informado feminino','Outros Feminino'] else 'Masculino' )
#Criando uma coluna apenas para 'Sexo' contendo Feminino e Masculino, incluindo os valores corretos para cada.
```

```python
#Em seguida, será definido na coluna "raca" os valores "branco(a)" e "negro(a)"
```

```python
def raca (df):
    if df['sexo_raça'] in ['Homem Branco', 'Mulher Branca']:
        return 'Branco'
    elif df['sexo_raça'] in ['Homem Negro', 'Mulher Negra']:
        return 'Negro'
    elif df['sexo_raça'] in ['Não informado masculino', 'Não informado feminino']:
        return 'Não informado'
    elif df['sexo_raça'] in ['Outros Feminino', 'Outros Masculino']:
        return 'Outros'
```

```python
df['raça'] = df.apply(raca, axis=1) #incluindo a função acima na coluna denominada "raça"
df
```

Eliminação das colunas

```python
df.drop(['sexo_raça',], axis = 1, inplace=True)
df
```

```python
df.drop(['Unnamed: 0'], axis = 1, inplace=True)
df
```

```python
df.rename(columns={"Sexo":"sexo","raça":"raca"}, inplace=True)
df
```
