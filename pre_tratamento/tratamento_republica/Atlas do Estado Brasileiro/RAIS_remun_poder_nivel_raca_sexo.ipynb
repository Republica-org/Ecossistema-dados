```python
import numpy as np
import pandas as pd
import os
df_wide = pd.read_csv('RAIS_remuneracao_poder_nivel_raca_sexo.csv', sep=';', encoding='latin-1')
df_wide
```

```python
df_wide.drop(df_wide.columns[df_wide.columns.str.contains('unnamed',case = False)],axis = 1, inplace = True)
df_wide
```

```python
df_long = df_wide.melt(id_vars =['esfera', 'poder', 'descricao_raca_ou_cor', 'genero'], 
        value_vars =['2004', '2005','2006', '2007','2008', '2009', '2010','2011','2012','2013','2014','2015','2016','2017','2018', '2019','2020'], 
        var_name ='Anos', 
        value_name ='Valores')
df_long
```

```python
df_long.rename(columns = {'esfera':'Esfera', 'poder':'Poder', 'descricao_raca_ou_cor' : 'Raça', 
                              'genero':'Gênero'}, inplace = True)
df_long
```

```python
df_long.to_csv('RAIS_remuneracao_poder_nivel_raca_sexo.csv')
```
